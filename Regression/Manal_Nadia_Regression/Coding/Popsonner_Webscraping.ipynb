```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.popsonner.com/p/top-1000-youtube-channels.html' 

response = requests.get(url)
```

```python
response.status_code
```

```python
page = response.text
page
```

```python
soup = BeautifulSoup(page,  "html.parser")
```

```python
print(soup.prettify())
```

```python
print(soup.find("table", id= "example").prettify())
```

```python
print(soup.find("tbody").tr.prettify())
```

```python
Pos_lst = []
Channel_lst = []
Views_lst = [] 
Subscriber_lst = [] 
Vidoes_lst = [] 
Published_lst = []

for i in (soup.find_all("tbody")):
    rows = i.find_all('tr')
    for tr in rows:
        cols = tr.find_all('td')
        for index, row in enumerate(cols):
            if index == 0:
                Pos_lst.append(row.text)
            elif index ==1: 
                Channel_lst.append(row.text)
            elif index ==2: 
                Views_lst.append(row.text)
            elif index ==3: 
                Subscriber_lst.append(row.text)
            elif index ==4: 
                Vidoes_lst.append(row.text)
            elif index ==5: 
                Published_lst.append(row.text)
            else:
                continue
```

```python
df = pd.DataFrame(list(zip(Pos_lst, Channel_lst, Views_lst, Subscriber_lst, Vidoes_lst, Published_lst)),
                    columns = ['Pos', 'Channel','Views','Subscriber', 'Vidoes', 'Published'])
```
